```python
import matplotlib.pyplot as plt
import numpy as np
import random
import math
```

```python
%matplotlib inline
```

```python
#функция для определения плотности выборочного нормального распределения
def norm(x, mu, sigm):
    return (math.exp(-(x-mu)**2 / 2 / sigm**2)/( (2 * math.pi)**0.5 * sigm))
```

# Распределение на 1ой выборке

```python
x = []
a = 1/1.05
#Задаём объём выборки
for i in range(1000):
    summ = 0
    #Задаём кол-во выборок
    for j in range(1):
        summ += (a)**random.expovariate(a)
    x.append(summ)
    
print("Дисперсия\t",  np.var(x))
print("Cреднеквадратичное отклонение\t", np.std(x))
print("Cреднее арифметическое\t",  np.mean(x))
#Cоздание гистограммы
plt.hist(x, bins = 'auto', normed = True)
#Наложение графика нормалльного распределения
plt.plot(sorted(x), [norm(val, np.mean(x), np.std(x)) for val in sorted(x)], color='black', lw=5)
plt.show()
```
