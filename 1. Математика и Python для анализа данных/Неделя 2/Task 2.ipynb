```python
import numpy as np
from scipy import linalg
```

```python
A = np.array([[1,1,1,1],[1,4,16,64],[1,10,100,1000],[1,15,225,3375]])
b = np.array([3.252,1.746,2.505,0.635])
```

```python
w = linalg.solve(A,b)
w
```

```python
%matplotlib inline
```

```python
from matplotlib import pylab as plt
x=np.arange(0,15.1,0.01)
y=w[0]+w[1]*x+w[2]*x*x+w[3]*x**3
plt.plot(x, y)
plt.show()
```

```python
with open('sol.txt', 'w') as f:
    s = ''
    for i in w:
        s += str(i) + ' '
    s = s.strip()
    f.write(s)
```
